```r
library(ISLR)
library(leaps)        # For best subset selection
library(glmnet)       # For ridge & lasso 
library(pls)          # For pcr & pls 
```

```r
names(Hitters)
dim(Hitters)
sum(is.na(Hitters$Salary))       # Missing salary for 59 players
Hitters = na.omit(Hitters)       # Data with incomplete cases removed
```

```r
# Best subset selection
## Outputs the best set of variables for each model size
## Asterisk: A given variable included in the corresponding model
## Default: To eight-variable model
reg_full_1 = regsubsets(x = Salary ~., data = Hitters)
summary(reg_full_1)

reg_full_2 = regsubsets(x = Salary ~., data = Hitters, nvmax = 19)
summary(reg_full_2)
```

```r
# Plot (Base plotting functions)
mod_summary = summary(reg_full_2)

par(mfrow = c(2, 2))

plot(mod_summary$rss, xlab = "Number Of Variables", ylab = "RSS", type = "l")
rss_min_ind = which.min(mod_summary$rss)
points(rss_min_ind, mod_summary$rss[rss_min_ind], col = "red", cex = 2, pch = 20)

plot(mod_summary$adjr2, xlab = "Number Of Variables", ylab = "Adjusted RSq", type = "l")
rsq_max_ind = which.max(mod_summary$adjr2)
points(rsq_max_ind, mod_summary$adjr2[rsq_max_ind], col = "red", cex = 2, pch = 20)

plot(mod_summary$cp, xlab = "Number Of Variables", ylab = "Cp", type = "l")
cp_min_ind = which.min(mod_summary$cp) 
points(cp_min_ind, mod_summary$cp[cp_min_ind], col = "red", cex = 2, pch = 20)

plot(mod_summary$bic, xlab = "Number Of Variables", ylab = "BIC", type = "l")
bic_min_ind = which.min(mod_summary$bic)
points(bic_min_ind, mod_summary$bic[bic_min_ind], col = "red", cex = 2, pch = 20)
```

```r
# Plot (Builtin package functions)
plot(reg_full_2, scale = "r2")
plot(reg_full_2, scale = "adjr2") 
plot(reg_full_2, scale = "Cp")
plot(reg_full_2, scale = "bic")
```

```r
# Forward & Backward stepwise selection
reg_fwd = regsubsets(Salary ~., data = Hitters, nvmax = 19, method = "forward")
summary(reg_fwd)

reg_bwd = regsubsets(Salary ~., data = Hitters, nvmax = 19, method = "backward")
summary(reg_bwd)
```

```r
# Model selection (Validation-set approach)
set.seed(1)
train = sample(c(TRUE, FALSE), nrow(Hitters), replace = TRUE)
test = (! train)

reg_full = regsubsets(Salary ~., data = Hitters[train, ], nvmax = 19)
valid_obs = model.matrix(Salary ~., data = Hitters[test, ])

val_err = rep(NA, 19)
for(i in 1:19) {
    # Extract coefficients (Optimal model of a given size)
    coef_i = coef(reg_full, id = i)
    # Make predictions (Matrix multiplication) 
    pred = valid_obs[, names(coef_i)] %*% coef_i
    val_err[i] = mean((Hitters$Salary[test] - pred) ^ 2)
}

which.min(val_err)  
```

```r
pred_fn = function(object, newdata, id, ...) {
    form = as.formula(object$call[[2]])
    matrix = model.matrix(form, newdata)
    coef_i = coef(object, id = id)
    xvars = names(coef_i)
    matrix[, xvars] %*% coef_i 
}
```

```r
# Model selection (Cross-validation approach)
set.seed(1)
k = 10

folds = sample(1:k, nrow(Hitters), replace = TRUE)
cv_err = matrix(data = NA, nrow = k, ncol = 19, dimnames = list(NULL, paste(1:19)))

# The ith cross-validation fold > The optimal j-variable model
for (i in 1:k) {
    reg_full = regsubsets(Salary ~., data = Hitters[(folds != i), ], nvmax = 19)
    for (j in 1:19) {
        pred_j = pred_fn(reg_full, Hitters[(folds == i), ], id = j)
        cv_err[i, j] = mean((Hitters$Salary[folds == i] - pred_j) ^ 2)
    }
}

# The CV error for the j-variable model
mean_cv_err = apply(cv_err, 2, mean)
print(mean_cv_err)
plot(mean_cv_err, type = 'b')
```

```r
# Ridge regression
x = model.matrix(Salary ~., Hitters)[, -1]         # Create x matrix (nobs * nvars)
y = Hitters$Salary                                 # Create y vector (nobs * 1)

grid = 10 ^ seq(10, -2, length = 100)              # Define decreasing sequence of lambda values (Null model (10^10) - Ridge model (10^-2) - OLS (0))
ridge_mod = glmnet(x, y, alpha = 0, lambda = grid) # Ridge (alpha = 0); Lasso (alpha = 1)
                                
dim(coef(ridge_mod))                               # Column: Lambda; Row: Predictor (For each lambda, obtain a vector of ridge coefs)
```

```r
# Property: The coef estimates should be smaller (L2 norm), when a large value of lambda is used
print(ridge_mod$lambda[10])                   # Large lambda 
print(sqrt(sum(coef(ridge_mod)[-1, 10] ^ 2))) # Small l2 norm 

print(ridge_mod$lambda[50])                   # Small lambda 
print(sqrt(sum(coef(ridge_mod)[-1, 50] ^ 2))) # Large l2 norm
```

```r
# Model selection (Validation-set approach)
set.seed(1)
train = sample(1:nrow(x), nrow(x)/2)
test = (-train)

ridge_mod = glmnet(x[train, ], y[train], alpha = 0, lambda = grid)

ridge_pred_1 = predict(ridge_mod, newx = x[test, ], s = 4)      # Lambda = 4 (Ridge model with moderate regularization)
mean((ridge_pred_1 - y[test]) ^ 2)
ridge_pred_2 = predict(ridge_mod, newx = x[test,], s = 1e10)    # Lambda = 10^10 (Null model with only intercept)
mean((ridge_pred_2 - y[test]) ^ 2)
ridge_pred_3 = predict(ridge_mod, newx = x[test,], s = 0)       # Lambda = 0 (OLS model with no regularization)
mean((ridge_pred_3 - y[test]) ^ 2)
```

```r
# Model selection (Cross-validation approach)
set.seed(1)
cv_ridge = cv.glmnet(x[train, ], y[train], alpha = 0)

optimal_lambda = cv_ridge$lambda.min # Lambda value that gives min mean cross-validated error
ridge_pred = predict(cv_ridge, newx = x[test, ], s = optimal_lambda) 
mean((ridge_pred - y[test]) ^ 2)     # Test error associated with selected lambda value  

plot(cv_ridge)
```

```r
# Lasso regression
## < Test MSE of null model and of least squares
## ~ Test MSE of ridge regression with lambda chosen by cross-validation
x = model.matrix(Salary ~., Hitters)[, -1]         
y = Hitters$Salary    
grid = 10 ^ seq(10, -2, length = 100)

set.seed(100)
train = sample(1:nrow(x), nrow(x)/2)
test = (-train)
cv_lasso = cv.glmnet(x[train, ], y[train], alpha = 1)

optimal_lambda = cv_lasso$lambda.min
lasso_pred = predict(cv_lasso, s = optimal_lambda, newx = x[test, ])
mean((lasso_pred - y[test]) ^ 2)  
  
plot(cv_lasso)
```

```r
lasso_mod = glmnet(x, y, alpha = 1, lambda = grid)
lasso_coef = predict(lasso_mod, type = "coefficients", s = optimal_lambda)[1:20, ]
print(lasso_coef)   # 13 of 19 variables' coefs = 0 
plot(lasso_mod)     # Coefs (Large value) ~ L1 norm (Large value = Small lambda) 
```

```r
# Principal components regression (PCR)
## The obtained model is more difficult to interpret ...
## Does not perform any kind of variable selection 
## Does not directly produce coefficient estimates
set.seed(2)
pcr_mod = pcr(Salary ~., data = Hitters, scale = TRUE, validation = "CV") # Standardize predictors prior to generating principal components
summary(pcr_mod)                                                          # Perform 10-fold cross-validation for each possible value of M
```

```r
# Validation statistics
## Mean squared error of prediction (MSEP)
## Root mean squared error of prediction (RMSEP) 
## 𝑅-squared (R2)
validationplot(pcr_mod, val.type = "MSEP")
```

```r
# Partial least squares (PLS)
## PCR attempts to maximize the amount of variance explained in the predictors
## PLS searches for directions that explain variance in both the predictors and the response
set.seed(1)
pls_mod = plsr(Salary ~., data = Hitters, scale = TRUE, validation = "CV")
summary(pls_mod)
```

```r
validationplot(pls_mod, val.type = "MSEP")
```
